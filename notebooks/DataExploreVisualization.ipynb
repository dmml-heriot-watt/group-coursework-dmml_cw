A bubble chart is a data visualization technique used to display data points in a three-dimensional space, where two axes represent attribute, and the third axis is represented by the size of the bubbles.

The provided code generates a bubble chart using the pandas and matplotlib libraries to visualize relationships between wine quality, alcohol content, and volatile acidity

The X-axis represents "Alcohol Content," which is a numerical attribute.

The Y-axis represents "Volatile Acidity," another numerical attribute.

Bubble Size: The size of each bubble represents the "Wine Quality" of the corresponding wine sample. Larger bubbles indicate higher wine quality, while smaller bubbles indicate lower quality. This allows you to quickly identify wines with different quality levels in the dataset.

Color Mapping: The color of each bubble is determined by the "Wine Quality" as well, and it follows a "coolwarm" color map. This provides an additional visual cue for quality, with cooler colors (e.g., blue) representing lower quality and warmer colors (e.g., red) representing higher quality.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Wine Quality dataset
df = pd.read_csv('../data/PreProcessed/MixWine.csv')

# Define the attribute for the bubble chart
x_attribute = 'alcohol'      
y_attribute = 'volatile acidity'
size_attribute = 'quality'          

# Create a bubble chart
plt.figure(figsize=(10, 8))
scatter = plt.scatter(df[x_attribute], df[y_attribute], s=df[size_attribute] * 20, c=df[size_attribute], cmap='coolwarm', alpha=0.8)
plt.title("Wine Quality vs. Alcohol Content and Volatile Acidity")
plt.xlabel("Alcohol Content")
plt.ylabel("Volatile Acidity")
plt.colorbar(label="Wine Quality")
plt.grid(True)
plt.show()
```

Heatmap is a visual representation of the correlations between pairs of attributes in a dataset. It uses colors to represent the strength and direction of the correlations. Here's how to interpret the elements of a correlation heatmap:

Color Scale: The heatmap uses a color scale to represent correlation values. In these color scales, positive correlations are typically shown in warm colors (e.g., red or orange), while negative correlations are in cool colors (e.g., blue).

Cells: Each cell in the heatmap represents the correlation between two attributes. The cell's color intensity indicates the strength of the correlation, with darker colors indicating stronger correlations. The color indicates the direction of the correlation (positive or negative).

Annotations: Numeric values inside the cells represent the correlation coefficients. These coefficients range from -1 to 1. A value of 1 indicates a perfect positive correlation, -1 indicates a perfect negative correlation, and 0 indicates no correlation. The closer the value is to 1 or -1, the stronger the correlation.

Diagonal Line: The diagonal line from the top-left to the bottom-right of the heatmap represents the correlation of each variable with itself. It will always be 1, as a variable is perfectly correlated with itself.

Heatmap in the context of the Wine Quality dataset:
Strong Positive Correlation: If two attributes have a strong positive correlation (e.g., a correlation coefficient close to 1), they tend to increase together. For example, if you find a dark red cell between "alcohol" and "density," it suggests that as alcohol content increases, density also tends to increase.

Strong Negative Correlation: A strong negative correlation (e.g., a correlation coefficient close to -1) indicates that two attributes tend to move in opposite directions. For instance, if "pH" has a dark blue cell with "fixed acidity," it suggests that as pH increases, fixed acidity tends to decrease.

No Correlation: If the cell is a pale color or close to white and the correlation coefficient is near 0, it suggests no significant linear relationship between the two attributes.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Wine Quality dataset
df = pd.read_csv('../data/PreProcessed/MixWine.csv')

# Select only the numeric columns for the correlation matrix
numeric_columns = df.select_dtypes(include=['number'])

# Compute the correlation matrix for numeric columns
correlation_matrix = numeric_columns.corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap of Numeric Attributes in Wine Quality Dataset")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Wine Quality dataset
df = pd.read_csv('MixWine.csv')

# Select only the numeric columns for box plots
numeric_columns = df.select_dtypes(include=['number'])

# Set the figure size
plt.figure(figsize=(15, 10))

# Create separate box plots for each numeric attribute
for i, col in enumerate(numeric_columns.columns):
    plt.subplot(3, 4, i + 1)  # Create a 3x4 grid of subplots
    sns.boxplot(x=df[col])
    plt.title(f"Box Plot of {col}")
    plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

plt.tight_layout()  # Ensure proper spacing between subplots
plt.show()
```
