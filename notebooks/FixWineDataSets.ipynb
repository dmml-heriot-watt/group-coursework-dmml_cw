### There are 1599 entries under Red wine dataset and 4898 in white. There is only 1 column in both that needs to be fixed

```python
###
###Function to fix Column name and return column name as an array, though this was not required
###
def FixColumnNames(df_wine):
    df = df_wine
    cols = df.columns[0].split(";")
    columns = []
    for c in cols:
        columns.append(c.strip('"').strip("'"))
    return columns
    
```

```python
##Get column names from the red wine dataset, as columns names are same for both red and white
columnNames = FixColumnNames(df_red) 
columnNames
```

```python
red_as_numpyarray = df_red.to_numpy() ## Get dateset to numpy array
red_as_numpyarray
```

```python
split_data_red = [row[0].split(';') for row in red_as_numpyarray] # To split each row by semicolon to get data for separate columns
```

```python
df_red_modified = pd.DataFrame(split_data_red, columns=columnNames)
df_red_modified
```

```python
white_as_numpyarray = df_white.to_numpy() ## Get dateset to numpy array
white_as_numpyarray
```

```python
split_data_white = [row[0].split(';') for row in white_as_numpyarray] # To split each row by semicolon to get data for separate columns
```

```python
df_white_modified = pd.DataFrame(split_data_white, columns=columnNames)
df_white_modified
```

```python
df_white_modified.to_csv("../data/PreProcessed/WhiteWineFixed.csv", index=False)
df_red_modified.to_csv("../data/PreProcessed/RedWineFixed.csv", index=False)
```
